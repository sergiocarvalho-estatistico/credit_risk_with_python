## Relav√¢ncia

Com base na teoria da informa√ß√£o concebida no final da d√©cada de 1940 e inicialmente desenvolvida para o desenvolvimento de 
scorecard, o Peso da Evid√™ncia (WOE) e o Valor da Informa√ß√£o (IV) v√™m ganhando cada vez mais aten√ß√£o nos √∫ltimos anos para usos 
como segmenta√ß√£o e redu√ß√£o de vari√°veis. Como o c√°lculo de WOE e IV requer o contraste entre ocorr√™ncia e n√£o ocorr√™ncia 
(geralmente denotado por 1 e 0), seu uso tem sido amplamente restrito √† an√°lise e previs√£o de resultados bin√°rios, 
como inadimpl√™ncia no pagamento, resposta a um marketing campanha, etc.

### WOE e IV para Vari√°veis Dependentes Cont√≠nuas. 

- Primeiro vamos propor um conjunto de f√≥rmulas alternativas de __WOE e IV__ para vari√°veis dependentes bin√°rias e examinar suas semelhan√ßas e diferen√ßas das f√≥rmulas originais. 
- Exploramos por que e como as f√≥rmulas alternativas podem ser adaptadas √† configura√ß√£o de vari√°veis dependentes cont√≠nuas.
- Resultados Esperados
    - Acelerar o processo de redu√ß√£o e sele√ß√£o de vari√°veis,  
    - Fornecer sugest√µes √∫teis para segmenta√ß√£o ou desenvolvimento de scorecard
    - Melhorar a imputa√ß√£o de valores ausentes, 
    - Vari√°veis de piso e teto, 
    - Cria√ß√£o de vari√°veis de classe, etc.


### IV e WOE para Vari√°veis Dependentes Bin√°rias

__Peso da Evid√™ncia (WOE)__ recodifica os valores de uma vari√°vel em categorias discretas e atribui a cada categoria um valor __WOE__ √∫nico no prop√≥sito de produzir as maiores diferen√ßas entre os recodificados. 

- Uma importante suposi√ß√£o aqui √© que a vari√°vel dependente deve ser bin√°ria para denotar a ocorr√™ncia e n√£o ocorr√™ncia de um evento. 
- No exemplo da an√°lise de risco, quando uma conta √© atual (boa) ou inadimplente (ruim), __WOE__ para qualquer segmento de clientes √© calculado da seguinte forma:

\begin{equation*}
WOE = \ln \left( \frac{\%ùêµùëéùëë_ùëñ}{\%ùê∫ùëúùëúùëë_ùëñ} \right) \times 100
\end{equation*}

![WOE](../imagens/woe1.png)

### Para entender melhor o WOE, gostar√≠amos de enfatizar o seguinte:

- __%Bad__ n√£o √© equivalente a __%Bad Rate__. 
- __%Bad__ √© a porcentagem de contas inadimplentes em uma faixa de pontua√ß√£o em rela√ß√£o a todas as contagens __Bad__. 
- O mesmo entendimento se aplica a __%Good__ no exemplo.
- N√£o importa se __%Good ou %Bad__ deve ser escolhido como numerador. 
- Se as duas medidas forem trocadas suas posi√ß√µes na divis√£o, o sinal de __WOE__ para cada faixa de pontua√ß√£o inverter√°, enquanto a magnitude permanecer√° inalterada. N√£o h√° impacto no c√°lculo subsequente de __IV__.
- A multiplica√ß√£o por 100 √© opcional.


Com base no __WOE__, __IV__ avalia o poder preditivo geral da vari√°vel. Portanto, pode ser usado para avaliar o poder preditivo entre vari√°veis concorrentes. O seguinte √© como o valor da informa√ß√£o √© calculado:

\begin{equation*}
IV = \sum_{i = 1}^{n} \left( (\%Bad_{i}-\%Good_{i}) \times \ln \left(\frac{\%Bad_{i}}{\%Good_{i}} \right) \right)
\end{equation*}

### Para infer√™ncias, gostar√≠amos de apontar o seguinte:

- Mesmo que as vari√°veis de __IV__ superior sejam consideradas mais preditivas, isso n√£o significa que elas est√£o prontas para entrar em um modelo de regress√£o diretamente. 
- A magnitude de __IV__ n√£o depende da monotonicidade da vari√°vel em rela√ß√£o para o resultado desejado. 
- Se duas bandas de pontua√ß√£o na Tabela 1 trocarem sua __% de taxa Bad__, isso resultar√° em nenhuma mudan√ßa no __IV__, mas uma vari√°vel como tal se torna menos √∫til para a regress√£o;
- Se o __WOE__ de uma vari√°vel num√©rica muda do maior para o menor (como na Tabela 1) ou vice-versa, geralmente sugere um boa monotonicidade e a vari√°vel torna-se um bom candidato para um modelo de regress√£o. As Figuras 1 e 2 ilustram este ponto com duas vari√°veis, uma com uma boa monotonicidade e outra sem;
- Se uma vari√°vel categ√≥rica mostra um alto IV, ela pode fornecer sugest√µes para segmenta√ß√£o.

E se a vari√°vel dependente for cont√≠nua, como receita, volume de vendas e valor da perda? Se um resultado n√£o for mais bin√°rio, as f√≥rmulas acima para WOE e IV n√£o se aplicam. No exemplo do volume de vendas, ainda podemos obter %Vendas (porcentagem de vendas em segmento sobre o volume total de vendas), mas %N√£oVendas √© totalmente inconceb√≠vel.

### WOE AND IV: ALTERMATIVE FORMULAE

Let‚Äôs examine an alternative set of formulae for the same risk analysis as in the previous section:

\begin{equation*}
WOE = \left[\ln \left(\frac{\%Bad_{i}}{\%Records_{i}} \right) \right] √ó 100 \\
IV = \sum_{i = 1}^{n} \left( (\%Bad_{i}-\%Records_{i}) \times \ln \left(\frac{\%Bad_{i}}{\%Records_{i}} \right) \right)
\end{equation*}


Em vez de comparar% bom com% ruim, a f√≥rmula acima substitui% bom por% de registros, que √© a porcentagem de contas em uma faixa de pontua√ß√£o sobre todas as contas, conforme ilustrado pela Tabela 2

## Redund√¢ncia - Crammer's V

No caso de vari√°veis cont√≠nuas na pr√°tica quando vamos modelar uma regress√£o logistica √© comum categorizarmos essas informa√ß√µes em faixas, segmentos com o intu√≠to de buscar uma certa estabilidade do nosso modelo. E para medir se vari√°veis categ√≥rios est√£o fortemente correlacionadas, ou seja, para medir se h√° redund√¢ncia essas informa√ß√µes utilizamos a __Correla√ß√£o de Crammer__.

\begin{equation*}
\sqrt \frac{\chi^2}{nt} \\[.8cm]
t = Minimum(r-1,c-1)
\end{equation*}

onde n √© o n√∫mero de registros e t √© o valor m√≠nimo entre o n√∫mero de categorias considerando as vari√°veis a serem testadas.

### C√°lculo de Correla√ß√µes (Crammer'V)

```python
import scipy.stats as ss
def cramers_v(confusion_matrix):
    chi2 = ss.chi2_contingency(confusion_matrix)[0]
    n = confusion_matrix.sum().sum()
    phi2 = chi2/n
    r,k = confusion_matrix.shape
    phi2corr = max(0, phi2-((k-1)*(r-1))/(n-1))
    rcorr = r-((r-1)**2)/(n-1)
    kcorr = k-((k-1)**2)/(n-1)
    return np.sqrt(phi2corr/min((kcorr-1),(rcorr-1)))
```
